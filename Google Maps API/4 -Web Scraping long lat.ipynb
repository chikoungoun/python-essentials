```python
import requests
import json
import csv
import time
```

```python
def cvs_write():
    csvfile = open('google_maps.csv','w')
    writer = csv.writer(csvfile,delimiter=',')
    writer.writerow(["name","lat","long"])
```

```python
api_key = 'YOUR API KEY'
```

** requires Google places API **

```python
url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?location=12.9538477,77.3507442&radius=20000&type=atm"
```

```python
r = requests.get(url + '&key=' + api_key) 
```

```python
response = r.json()
```

```python
response
```

```python
for i in response['results']:
    print(i['geometry']['location'])
```
