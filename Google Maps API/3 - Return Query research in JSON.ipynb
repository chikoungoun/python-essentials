# 3 - Search Places in a Map

```python
import requests
import json
```

```python
api_key = 'Your API'
```

** requires Google places API **

```python
url = "https://maps.googleapis.com/maps/api/place/textsearch/json?"
```

```python
query = "pharmacie rabat"
```

```python
r = requests.get(url + 'query=' + query +'&key=' + api_key) 
```

```python
x = r.json()
```

```python
x
```

```python
y = x['results']
```

```python
for i in range(len(y)):
    print(y[i]['name'])
```
